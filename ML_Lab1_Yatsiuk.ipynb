# ***Імпортуємо необхідні бібліотеки для роботи***

```python
import numpy as ny
import matplotlib.pyplot as ppt
from sklearn.cluster import KMeans
import statistics
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

# ***Підготуємо довільну вибірку даних***

```python
X, y_true = make_blobs(n_samples=1000, centers=3, cluster_std=1, random_state=1)
ppt.scatter(X[:, 1], X[:, 0], s=25);
```

# ***Проведемо необхідний аналіз до сформованих кластерів*** 

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
ppt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 20, c = 'orange', label = 'Cluster1')
print('Orange | Arithmetic mean: ', (statistics.mean(X[y_kmeans == 0, 1])), '| Median:', (statistics.median(X[y_kmeans == 0, 1])))
ppt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 20, c = 'blue', label = 'Cluster2')
print('Blue | Arithmetic mean: ', (statistics.mean(X[y_kmeans == 1, 1])), '| Median:', (statistics.median(X[y_kmeans == 1, 1])))
ppt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 20, c = 'green', label = 'Cluster3')
print('Green | Arithmetic mean: ', (statistics.mean(X[y_kmeans == 2, 1])), '| Median:', (statistics.median(X[y_kmeans == 2, 1])))
ppt.show()
```
