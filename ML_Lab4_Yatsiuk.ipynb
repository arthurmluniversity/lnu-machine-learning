# ***Імпортуємо необхідні бібліотеки для роботи***

```python
import matplotlib.pyplot as plt
import scipy as sc
import numpy as np
from scipy.misc import derivative
import unittest
```

# ***Побудова графіку функції***

```python
x = np.arange(-5, 5.01, 0.01)
plt.figure(figsize=(10, 5))
plt.plot(x, ((2*x+3*x)-5*x)+pow(x,2), label=r'$f_1(x)=((2x+3x)-5x))+x^2$')
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$f(x)$', fontsize=14)
plt.grid(True)
plt.legend(loc='best', fontsize=12)
plt.show()
```

# ***Знаходимо похідну за допомогою скінченних різниць, а для тестування використовуємо бібліотеку “derivative”***

```python
def f(x):
    return 2*x
derivative(f, 1.0, dx=0.01)
```

# ***Реалізація градієнтного спуску для пошуку глобального мінімуму функції.***

```python
ma = 0 
min = 6 
gamma = 0.01 
precision = 0.00001

def f_derivative(x):
    return 2*x

while abs(min - ma) > precision:
    ma = min
    min = ma - gamma * f_derivative(ma)

print("Локальний мінімум має місце в", min)
```

# ***Візуалізація результату градієнтного спуску на графіку, як друга легенда відповідно до 1-го пункту завдань.***


```python
y=ma - gamma * f_derivative(ma)
fig, crashed = plt.subplots()
crashed.plot(x, ((2*x+3*x)-5*x)+pow(x,2), label = r'$f_1(x)=((2x+3x)-5x))+x^2$')
crashed.scatter(y, y, c="r", label = 'Локальний мінімум')
crashed.legend(
    fontsize = 10,
          ncol = 2,    
          facecolor = 'yellow',    
          edgecolor = 'r',    
          title = 'Візуалізація результату градієнтного спуску на графіку',    
          title_fontsize = '10'    
)
plt.grid(True)
fig.set_figheight(5)
fig.set_figwidth(8)
plt.show()
```

# ***Виведення результату (глобального мінімуму) градієнтного спуску та візуалізація на графіку у вигляді маркера.***

```python
min = plt.figure().add_subplot(111, projection='3d')
plt.plot(x, ((2*x+3*x)-5*x)+pow(x,2), label='$f_1(x)=((2x+3x)-5x)+x^2$')
min.scatter(0.0003202000669438543, 0.0003202000669438543,  color='red')
    
plt.show()
```

# ***Тестування методу.***

```python
def func(x):
    return ((2*x+3*x)-5*x)+pow(x,2)
def pohidna(x, h):
    return (func(x+h) - func(x-h)) / (2 * h)
class TestNotebook(unittest.TestCase):
    
    def test_pohidna(self):
        self.assertAlmostEqual(pohidna(1.0, 0.01), 2)
    def test_pohidna2(self):
        self.assertAlmostEqual(derivative(f, 1.0, dx=0.01), 2)
        

unittest.main(argv=[''], verbosity=2, exit=False)
```
