# ***Імпортуємо та виведемо на екран дані з файлу "nba.csv".***

```python
NBA = pd.read_csv('nba.csv')
pd.set_option('display.max_columns', 70)
print(NBA)
```

# ***Проведемо аналіз на цілісність даних та виконаємо конвертацію даних для стовпців “Height”, “Weight”.***

```python
nba_g = NBA.dropna()
print(nba_g)
```

```python
def height_m(x):
  x = str(x).split('-')
  return round(int(x[0]) + int(x[1]) / 12) * 0.3048
```

```python
nba_g['Height'] = nba_g['Height'].transform(height_m)
```

```python
nba_g['Weight'] = nba_g['Weight'].apply(lambda x:x * 0.45359237)
```

```python
print(nba_g)
```

# ***Реалізуємо візуалізацію для валідованих та конвертованих даних "employees.csv" у форматі графіків.***

```python
g1 = employees_g.groupby(by="Team", dropna=False).max()
print(g1)
```

```python
g2 = employees_g.groupby(by="Gender", dropna=False).min()
print(g2)
```

```python
g02 = employees_g.groupby(by="Senior Management", dropna=False).min()
print(g02)
```

```python
g1.plot()
```

```python
g2.plot.pie(y='Bonus %', figsize=(4, 4))
g2.plot.pie(y='Salary', figsize=(4, 4))
```

```python
g02.plot.pie(y='Salary', figsize=(4, 4))
g02.plot.pie(y='Bonus %', figsize=(4, 4))
```

# ***Реалізуємо візуалізацію для валідованих та конвертованих даних "nba.csv" у форматі графіків.***

```python
n1 = nba_g.groupby(by="Team", dropna=False).min()
print(n1)
```

```python
n1.plot.bar(x='Number', y='Age',figsize=(10, 10))
```

```python
n1.plot.barh(x='Height', y='Weight', figsize=(10, 10))
```

```python
n2 = nba_g.groupby(by="College", dropna=False).mean()
print(n2)
```

```python
n2.plot.bar(x='Number', y='Age',figsize=(10, 10))
```

```python
n2.plot.barh(x='Weight', y='Height', figsize=(10, 10))
```

# ***Експортуємо валідовані та конвертовані дані у файли "employees_ex.csv", "nba_ex.xlsx".***

```python
g1.to_csv('employees_ex.csv', index=False)
n1.to_excel('nba_ex.xlsx', index=False)
```
